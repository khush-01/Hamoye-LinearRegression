# Importing libraries and the dataset

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler

from sklearn.linear_model import Ridge, Lasso, LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
quiz_data = pd.read_csv('energydata_complete.csv')
quiz_data.head()
```

```python
quiz_data.drop(['lights'], axis=1,inplace=True)
quiz_data.drop(['date'], axis=1, inplace=True)
```

**Preprocessing**

```python
scaler = MinMaxScaler()
normalized_features = pd.DataFrame(scaler.fit_transform(quiz_data), columns=quiz_data.columns)

features = normalized_features.drop(['Appliances'], axis=1)
target = normalized_features['Appliances']
```

```python
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size = 0.3, random_state = 42)
X_train.head()
```

```python
print('Question 12\n')

Xtrain = X_train[['T2']]
Ytrain = X_train[['T6']]

Xtest = X_test[['T2']]
Ytest = X_test[['T6']]

model = LinearRegression()
model.fit(Xtrain, Ytrain)
predict = model.predict(Xtest)

r2_value = round(r2_score(Ytest, predict), 2)
print('R2 Score =', r2_value)
```

```python
#Linear Regression Model for 13-17

linear_regression = LinearRegression()
linear_regression.fit(X_train, y_train)
y_predicted = linear_regression.predict(X_test)
```

```python
print('Question 13\n')

mae_value = round(mean_absolute_error(y_test, y_predicted), 2)
print('MAE =', mae_value)
```

```python
print('Question 14\n')

rss_value = round(np.sum(np.square(y_test - y_predicted)), 2)
print('RSS =', rss_value)
```

```python
print('Question 15\n')

rmse_value = round(np.sqrt(mean_squared_error(y_test, y_predicted)), 3)
print('RMSE =', rmse_value)
```

```python
print('Question 16\n')

r2_value = r2_score(y_test, y_predicted)
print('Coefficient of Determination =', round(r2_value, 2))
```

```python
print('Question 17\n')

def get_weights(model, feature, column_name):
  weights = pd.Series(model.coef_, feature.columns).sort_values()
  weights_dataframe = pd.DataFrame(weights).reset_index()
  weights_dataframe.columns = ['Features', column_name]
  weights_dataframe[column_name].round(3)
  return weights_dataframe

linear_weights = get_weights(linear_regression, X_train, 'Linear_weight')
print('Lowest & Highest:')
linear_weights.iloc[[0, -1], :]
```

```python
print('Question 18\n')

ridge = Ridge()
ridge.fit(X_train, y_train)
predicted = ridge.predict(X_test)

ridge_rmse_value = round(np.sqrt(mean_squared_error(y_test, predicted)), 3)

print(f'The RMSE score for the rigde model with default aplha value when evaluated on the test set is: {ridge_rmse_value}')


ridge_reg = Ridge(alpha=0.4)
ridge_reg.fit(X_train, y_train)
new_predicted = ridge_reg.predict(X_test)

ridge_rmse_new = round(np.sqrt(mean_squared_error(y_test, new_predicted)), 3)

print(f'The RMSE score for the rigde model with new aplha value when evaluated on the test set is: {ridge_rmse_new}')
```

```python
print('Question 19\n')

lasso_regression = Lasso(alpha=0.001)
lasso_regression.fit(X_train, y_train)
lasso_predicted = lasso_regression.predict(X_test)

def get_weights(model, feature, column_name):
  weights = pd.Series(model.coef_, feature.columns).sort_values()
  weights_dataframe = pd.DataFrame(weights).reset_index()
  weights_dataframe.columns = ['Features', column_name]
  weights_dataframe[column_name].round(3)
  return weights_dataframe

lasso_weights = get_weights(lasso_regression, X_train, 'Lasso_weight')
non_zero_weights = lasso_weights[lasso_weights['Lasso_weight'] != 0]

print('No of features =', len(non_zero_weights))
```

```python
print('Question 20\n')

new_rmse_value = round(np.sqrt(mean_squared_error(y_test, lasso_predicted)), 3)
print('New RMSE =', new_rmse_value)
```
